```python
from typing import TypedDict, List
from langgraph.graph import StateGraph, START, END
```

```python
# agent state
class AgentState(TypedDict):

    name: str
    values : List[int]
    result : str

# Node
def process(state: AgentState) -> AgentState:

    state["result"] = f"Hi {state["name"]}. your values sum is: {sum(state["values"])}"

    return state


```

```python

graph = StateGraph(AgentState)

#to add a node

graph.add_node("process", process)

graph.add_edge(START, "process")
graph.add_edge("process", END)
```

```python
app = graph.compile()
```

```python
result = app.invoke({"name": "Babar", "values": [1,2,3,4,5]})
result
```

```python
result["result"]
```

## task

- create a graph where you can pass in a single list of integers along with name and an operation . 
- if the operater is '+', you add the elements, if its a '*' you multiply the elements, all wihin the same node.

- input: {"name" : "babar". "values": [1,2,3], "Operation": "*"}
- output: "Hi babar, you anwer is 6."

```python
import math
```

```python

class Agentstate(TypedDict):

    name: str
    values: List[int]
    operator : str
    message: str



# Define your node function here
def process_operation(state: Agentstate) -> Agentstate:
    
    operator = state["operator"]
    
    if operator == "+":
        ans = sum(state["values"])

    elif operator == "*":
        ans = math.prod(state["values"])
        
    state["message"] = f"Hi {state['name']}, your answer is {ans}."

    return state
    
```

```python
graph = StateGraph(Agentstate)

graph.add_node("process_node", process_operation)

graph.add_edge(START, "process_node")
graph.add_edge("process_node", END)

```

```python
app = graph.compile()
```

```python
result = app.invoke({"name": "Babar", "values": [1,2,3,4,5], "operator": "*"})

```
